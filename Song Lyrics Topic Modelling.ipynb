Word data cleaning

```python
with open('genre_sample_tracks.pickle', 'rb') as f:
    sample_tracks = pickle.load(f)

# merge genre information onto lyrics data
sample_tracks = sample_tracks.merge(genre_sample,on='track_id',how='left',validate='m:1')

# update stemmed words into their full version
sample_tracks = sample_tracks.merge(word_convert,on='word',how='left',validate='m:1',indicator = True)
# print("displaying 5 random words to be updated from their stems")
# print(sample_tracks[(sample_tracks['word']!=sample_tracks['full word'])&(sample_tracks['_merge']=='both')].sample(n=5)[['word','full word']])
sample_tracks.loc[sample_tracks['_merge']=='both','word'] = sample_tracks['full word'] 
sample_tracks = sample_tracks.drop(columns=['full word','_merge'])

# manual cleaning - censoring
sample_tracks.loc[sample_tracks['word'].str.contains("^nigga"),'word'] = 'n-word'

# manual cleaning - beginning of word contracted 
sample_tracks.loc[sample_tracks['word']== "'bout",'word'] = 'about'
sample_tracks.loc[(sample_tracks['word']== "'cause") | (sample_tracks['word']=="'cos"),'word'] = 'because' 
sample_tracks.loc[sample_tracks['word']== "'em",'word'] = 'them'
sample_tracks.loc[sample_tracks['word']== "'fore", 'word'] = 'before'

# manual cleaning - end of word contracted 
sample_tracks.loc[sample_tracks['word']== 'yea','word'] = 'yeah'
sample_tracks.loc[sample_tracks['word']== "gon'", 'word'] = 'gonna'
sample_tracks.loc[sample_tracks['word']== "po'", 'word'] = 'po'
sample_tracks.loc[sample_tracks['word']== "piu'", 'word'] = 'piu'
sample_tracks.loc[sample_tracks['word']== "ol'", 'word'] = 'old'
sample_tracks.loc[sample_tracks['word']== "lil'", 'word'] = 'little'
sample_tracks.loc[sample_tracks['word']== "perche'", 'word'] = 'perche'
sample_tracks.loc[sample_tracks['word'].str.contains("'$"), 'word'] = sample_tracks['word'].str.replace("'$",'g')
```

#### Consolidate data after cleaning individual terms (Pickling here)

```python
sample_tracks_summed = sample_tracks.groupby(['track_id','mxm_tid','word','is_test','tag']).sum().reset_index()

with open('sample_tracks_summed.pickle', 'wb') as f:
    pickle.dump(sample_tracks_summed, f, protocol=4)
```

#### Add stop words here

```python
with open('sample_tracks_summed.pickle', 'rb') as f:
    sample_tracks_summed = pickle.load(f)
    
custom_sw = ['would','because','cause','oh','ooh',"'til", 'gonna', 'oooh',
             'eh', 'hey', 'uhh', 'cha','ca', 'na', 'yeah', 'ah', 'gotta', 'us',
             'ba', 'uh', "c'mon","y'all","n-word",'baby', 'shit', 'fuck', 'like',
             'got','get','gotta'
            ]

nostops = sample_tracks_summed.copy()
nostops = nostops[
    ~(nostops['word'].isin(stopwords.words('english'))) &
    ~(nostops['word'].isin(stopwords.words('french'))) &
    ~(nostops['word'].isin(stopwords.words('german'))) &
    ~(nostops['word'].isin(stopwords.words('spanish'))) & 
    ~(nostops['word'].isin(stopwords.words('italian'))) & 
    (nostops['word'].str.contains('[^0-9\W]')) &
    ~(nostops['word'].isin(custom_sw))
    ]
```

```python
# Track level data
total_counts = sample_tracks_names.groupby(['track_id','artist name','title'])['count'].sum().reset_index()
total_counts = total_counts.rename(columns={'count':'total word count'})

diff_words = sample_tracks_names.groupby(['track_id'])['word'].count().reset_index()
diff_words = diff_words.rename(columns={'word':'num diff words'})

word_freq_index = total_counts.merge(diff_words,on='track_id',how='left',validate='1:1')
word_freq_index['word freq index'] = word_freq_index['total word count'] / word_freq_index['num diff words']  

# look at songs with low unique words and high word counts 
high_freq_tracks = word_freq_index[((word_freq_index['word freq index'] >=5)
                & (word_freq_index['num diff words'] <=10))].sort_values(by='word freq index',ascending=False)

print(f"Number of high frequency tracks is {len(high_freq_tracks)}")
high_freq_tracks.head()
```

# Topic Modelling

```python
model_data = sample_wide 
vocab = sample_wide.filter(regex=' word$',axis=1).columns.tolist() # list of words in order of column index
```

## NMF

```python
# fit the transfomed content with NMF

n_topics = 9
nmf = NMF(n_components=n_topics,max_iter=1000)
```

```python
# Count vectorized
nmf.fit(model_data.filter(regex=' word$',axis=1))
```

```python
# display the result
from collections import defaultdict

top_words_dict = {}
top_weights_dict = {}
top_words_count = defaultdict(int)
for index,topic in enumerate(nmf.components_):
    n = 11
    print(f"The top {n-1} words for topic # {index}")
    top_words =[vocab[i].replace(' word','') for i in topic.argsort()[:-1*n:-1]]
    top_weights = [topic[i] for i in topic.argsort()[:-1*n:-1]]
    print(top_words)
    print("\n")
    top_words_dict[index] = top_words
    top_weights_dict[index] = top_weights
    for word in top_words:
        top_words_count[word]+=1
```

```python
all_topic_keys = pd.DataFrame()
for i,topic_i in enumerate([1,2,6,8,0]):
    topic_keys = pd.DataFrame(top_weights_dict[topic_i],columns=['Weight'])
    topic_keys['Word'] = pd.Series(top_words_dict[topic_i])
    word_counts = sample_no_repetitives[sample_no_repetitives['word'].isin(top_words_dict[topic_i])].groupby('word').sum()['count'].reset_index()
    topic_keys = topic_keys.merge(word_counts,left_on='Word',right_on='word',how='left',validate='1:1').drop(columns=['word'])
    topic_keys = topic_keys.rename({'count':'Count'}) 
    topic_keys['Topic'] = i
    all_topic_keys = all_topic_keys.append(topic_keys)
    
all_topic_keys = all_topic_keys.reset_index()
# all_topic_keys
```
